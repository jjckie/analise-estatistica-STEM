```python
import pandas as pd 
df = pd.read_csv('stem_database.csv')
df.head()
```

```python
df.info()
df.describe()
```

Na poxima sessão, são analisadas medidas de tendencias central(media, mediana e moda) para a coluna do ano de 2023 (quantidade geral de estudantes de stem)

```python

media = df['2023'].mean()
mediana = df['2023'].median()
moda = df['2023'].mode()[1]
print(media)
print(mediana)
print(moda)
```

**Média (19,14%)**
A média representa o valor médio entre todos os países analisados. Isso indica que, em 2023, 19,14% dos formandos, em média, se graduaram em cursos da área STEM.

**Mediana (22,40%)**
A mediana é o valor que fica exatamente no meio da distribuição dos dados. Neste caso, metade dos países teve menos de 22,40% de formandos em STEM, e a outra metade teve mais.

**Moda (17,84%)**
A moda é o valor que mais aparece entre os dados. Neste conjunto, 17,84% foi a porcentagem mais frequente.

Na poxima sessão, são analisadas medidas de dispersão (desvio padrão, variância e amplitude) para a coluna do ano de 2023 (quantidade geral de estudantes de stem)

```python
desvio = df['2023'].std() 
variancia = df['2023'].var()  
amplitude = df['2023'].max() - df['2023'].min()  

print(desvio)
print(variancia)
print(amplitude)
```

**Desvio Padrão: 14,00%**
O desvio padrão indica o quanto os dados variam em relação à média. Com um desvio de 14%, podemos afirmar que os países têm grandes diferenças entre si nas taxas de graduação em STEM.

**Variância: 195,98**
A variância é o desvio padrão ao quadrado. Neste caso, o valor 195,98 confirma que os dados estão bem espalhados em torno da média.

**Amplitude: 31,74%**
A amplitude é a diferença entre o maior (31,74%) e o menor (0%) valor da coluna. Isso mostra que existe uma grande desigualdade entre os países analisados, com alguns tendo muitos e outros nenhum graduado em STEM em 2023.

A seguir, serão analisadas as colunas dos anos de 1998 (início da série de dados) e 2010.

```python
media_1998 = df['1998'].mean()
mediana_1998 = df['1998'].median()
moda_1998 = df['1998'].mode()[0] 
desvio_1998 = df['1998'].std()
variancia_1998 = df['1998'].var()
amplitude_1998 = df['1998'].max() - df['1998'].min()

print("Ano 1998")
print("Média:", media_1998)
print("Mediana:", mediana_1998)
print("Moda:", moda_1998)
print("Desvio padrão:", desvio_1998)
print("Variância:", variancia_1998)
print("Amplitude:", amplitude_1998)
```

```python
media_2010 = df['2010'].mean()
mediana_2010 = df['2010'].median()
moda_2010 = df['2010'].mode()[0]
desvio_2010 = df['2010'].std()
variancia_2010 = df['2010'].var()
amplitude_2010 = df['2010'].max() - df['2010'].min()

print("\nAno 2010")
print("Média:", media_2010)
print("Mediana:", mediana_2010)
print("Moda:", moda_2010)
print("Desvio padrão:", desvio_2010)
print("Variância:", variancia_2010)
print("Amplitude:", amplitude_2010)
```

Analisando os anos de 1998, 2010 e 2023, nota-se que a média de formandos em cursos STEM teve pequenas variações: 18,26% em 1998, 20,27% em 2010 e 19,14% em 2023, sem um crescimento contínuo. A mediana se manteve próxima nos três anos: 22,26% em 1998, 20,53% em 2010 e 22,40% em 2023, indicando certa estabilidade na posição central dos dados. A moda foi 0,0 nos anos de 1998 e 2010, sugerindo que muitos países não tinham graduados em STEM nesses períodos. Em termos de dispersão, houve uma queda significativa na variância de 375,17 em 1998 para 55,48 em 2010, indicando menor desigualdade naquele período, mas um aumento novamente em 2023, com variância de 195,98, sugerindo que as diferenças entre os países voltaram a crescer.
