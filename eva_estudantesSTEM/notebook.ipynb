```python
import pandas as pd 
df = pd.read_csv('stem_database.csv')
df.head()
```

```python
df.info()
df.describe()
```

Na poxima sesso, so analisadas medidas de tendencias central(media, mediana e moda) para a coluna do ano de 2023 (quantidade geral de estudantes de stem)

```python

media = df['2023'].mean()
mediana = df['2023'].median()
moda = df['2023'].mode()[1]
print(media)
print(mediana)
print(moda)
```

**Mdia (19,14%)**
A mdia representa o valor mdio entre todos os pases analisados. Isso indica que, em 2023, 19,14% dos formandos, em mdia, se graduaram em cursos da rea STEM.

**Mediana (22,40%)**
A mediana  o valor que fica exatamente no meio da distribuio dos dados. Neste caso, metade dos pases teve menos de 22,40% de formandos em STEM, e a outra metade teve mais.

**Moda (17,84%)**
A moda  o valor que mais aparece entre os dados. Neste conjunto, 17,84% foi a porcentagem mais frequente.

Na poxima sesso, so analisadas medidas de disperso (desvio padro, varincia e amplitude) para a coluna do ano de 2023 (quantidade geral de estudantes de stem)

```python
desvio = df['2023'].std() 
variancia = df['2023'].var()  
amplitude = df['2023'].max() - df['2023'].min()  

print(desvio)
print(variancia)
print(amplitude)
```

**Desvio Padro: 14,00%**
O desvio padro indica o quanto os dados variam em relao  mdia. Com um desvio de 14%, podemos afirmar que os pases tm grandes diferenas entre si nas taxas de graduao em STEM.

**Varincia: 195,98**
A varincia  o desvio padro ao quadrado. Neste caso, o valor 195,98 confirma que os dados esto bem espalhados em torno da mdia.

**Amplitude: 31,74%**
A amplitude  a diferena entre o maior (31,74%) e o menor (0%) valor da coluna. Isso mostra que existe uma grande desigualdade entre os pases analisados, com alguns tendo muitos e outros nenhum graduado em STEM em 2023.

A seguir, sero analisadas as colunas dos anos de 1998 (incio da srie de dados) e 2010.

```python
media_1998 = df['1998'].mean()
mediana_1998 = df['1998'].median()
moda_1998 = df['1998'].mode()[0] 
desvio_1998 = df['1998'].std()
variancia_1998 = df['1998'].var()
amplitude_1998 = df['1998'].max() - df['1998'].min()

print("Ano 1998")
print("Mdia:", media_1998)
print("Mediana:", mediana_1998)
print("Moda:", moda_1998)
print("Desvio padro:", desvio_1998)
print("Varincia:", variancia_1998)
print("Amplitude:", amplitude_1998)
```

```python
media_2010 = df['2010'].mean()
mediana_2010 = df['2010'].median()
moda_2010 = df['2010'].mode()[0]
desvio_2010 = df['2010'].std()
variancia_2010 = df['2010'].var()
amplitude_2010 = df['2010'].max() - df['2010'].min()

print("\nAno 2010")
print("Mdia:", media_2010)
print("Mediana:", mediana_2010)
print("Moda:", moda_2010)
print("Desvio padro:", desvio_2010)
print("Varincia:", variancia_2010)
print("Amplitude:", amplitude_2010)
```
